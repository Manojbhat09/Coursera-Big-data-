```python
sqlContext = SQLContext(sc)
df = sqlContext.read.load('file:///home/cloudera/Downloads/coursera-master/big-data-4/daily_weather.csv', 
                          format='com.databricks.spark.csv', 
                          header='true',inferSchema='true')
df.columns
```

```python
featureColumns = ['air_pressure_9am','air_temp_9am','avg_wind_direction_9am','avg_wind_speed_9am',
        'max_wind_direction_9am','max_wind_speed_9am','rain_accumulation_9am',
        'rain_duration_9am']
```

```python
df2 = sqlContext.read.load('file:///home/cloudera/Downloads/coursera-master/big-data-4/daily_weather.csv', 
                          format='com.databricks.spark.csv', 
                          header='true',inferSchema='true')
```

```python
df2.toPandas()
```

```python
df = df.drop('number')
```

```python
df = df.na.drop()
```

```python
df.count()
```

```python
len(df.columns)
```

```python
binf = Binarizer(threshold= 24.999 , inputCol= 'relative_humidity_3pm', outputCol= 'label')
binDf = binf.transform(df)
```

```python
binDf.select('relative_humidity_3pm', 'label').show()
```

```python
assembler = VectorAssembler(inputCols=featureColumns, outputCol='features')
asmbled = assembler.transform(binDf)
asmbled.count()
```

```python
(trainingDat, testDat) = asmbled.randomSplit([0.8,0.2], seed = 13234)
```

```python
print(trainingDat.count())
print(testDat.count())
print(asmbled.count())
```

```python
asmbled.toPandas().transpose()
```

```python
dt = DecisionTreeClassifier(labelCol="label", featuresCol="features", maxDepth=5, minInstancesPerNode=20, impurity="gini")
```

```python
dt.getMinInfoGain
```

```python
[dt]
```

```python
pipeline = Pipeline(stages= [dt])
```

```python
model = pipeline.fit(trainingDat)
```

```python
predictions = model.transform(testDat)
```

```python
print(predictions.select('relative_humidity_3pm', 'label').show())
print(predictions.select('prediction', 'label').show())
```

```python
predictions.select('prediction', 'label').write.save(path = "file:///home/cloudera/Downloads/coursera-master/big-data-4/predictions.csv", format ="com.databricks.spark.csv", header="true")
```
